```python
from scipy.optimize import curve_fit
def func(x,a,b,c):
    return a*np.exp(-b*x)+c
xdata=np.linspace(0,10,100)
y=func(xdata, 2.5,1.3,0.5)
y_noise=0.5*np.random.rand(100)
ydata=y+y_noise
plt.plot(xdata,ydata,'o')
popt,pcov=curve_fit(func,xdata,ydata)
#print(popt)
plt.plot(xdata, func(xdata,popt[0],popt[1],popt[2]))
```

```python
def sigmoid(x,L,x0,k):
    return L/(1+np.exp(-k*(x-x0)))
xx=np.array(range(len(dfc.index)))
yy=dfc.Nepal
plt.plot(xx,yy,'o','g',label='Cases:data')
mean=np.mean(yy)
sig=np.std(yy)
p0=[1,mean,1]

popt,pcov=curve_fit(sigmoid,xx,yy)
plt.plot(xx,sigmoid(xx,*popt),'r',label='Fit')
plt.legend()
```

```python
plt.figure(figsize=(10,7))
plt.plot(xx, yy, 'o',label='Cases: data',lw=3,color='g')

xnew = np.array( range(600) )
plt.plot(xnew, sigmoid(xnew, *popt), label='Prediction', lw=3,color='r')
plt.xlabel('Days from starting', color='k',fontsize=18)
plt.ylabel('Number of cases', color='k',fontsize=18)
plt.legend(fontsize=16,loc='best',frameon=False)
plt.title('Covid Cases of Nepal', fontsize=22, color='r')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#plt.savefig('CovidNepal.jpg')

```
