# how to use the date column as index column

```python
A = dfc.Afghanistan.values.reshape(-1,1)
I = dfc.India.values.reshape(-1,1)
N = dfc.Nepal.values.reshape(-1,1)
P = dfc.Pakistan.values.reshape(-1,1)
B = dfc.Bangladesh.values.reshape(-1,1)
S = dfc['Sri Lanka'].values.reshape(-1,1)
M = dfc.Maldives.values.reshape(-1,1)
Bh = dfc.Bhutan.values.reshape(-1,1)
```

```python
scaled_A = preprocessing.MinMaxScaler().fit(A).transform(A)
scaled_N = preprocessing.MinMaxScaler().fit(N).transform(N)
scaled_I = preprocessing.MinMaxScaler().fit(I).transform(I)
scaled_P = preprocessing.MinMaxScaler().fit(P).transform(P)
scaled_B = preprocessing.MinMaxScaler().fit(B).transform(B)
scaled_S = preprocessing.MinMaxScaler().fit(S).transform(S)
scaled_M = preprocessing.MinMaxScaler().fit(M).transform(M)
scaled_Bh = preprocessing.MinMaxScaler().fit(Bh).transform(Bh)
```

```python
plt.figure(figsize=[10,10])
plt.plot(dfc.index,scaled_N,'-.',color='g',lw=3,label='Nepal(80,000)')
plt.plot(dfc.index,scaled_I,'-.', color='r',lw=4,label='India')
plt.plot(dfc.index,scaled_P,'-', color='b',lw=4,label='Pakistan')
plt.plot(dfc.index,scaled_B, color='r',lw=4,label='Bangladesh')
plt.plot(dfc.index,scaled_A, '-.',color='k',lw=3,label='Afghanistan')
plt.plot(dfc.index,scaled_S, color='m',lw=3,label='Sri Lanka')
plt.plot(dfc.index,scaled_M, '-.',color='b',lw=3,label='Maldives')
plt.plot(dfc.index,scaled_Bh,'-', color='k',lw=3,label='Bhutan')
plt.legend(fontsize=20)
plt.title('COVID-19 Comparison in SAARC', fontsize=20,weight='bold')
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Date', fontsize=20, weight='bold')
plt.ylabel('Scaled Cases', fontsize=20,weight='bold')
#plt.savefig('SAARCO.png')
```

```python
US = dfc['United States'].values.reshape(-1,1)
I = dfc.India.values.reshape(-1,1)
Br = dfc.Brazil.values.reshape(-1,1)
R = dfc.Russia.values.reshape(-1,1)
C = dfc.Colombia.values.reshape(-1,1)
Pe = dfc.Peru.values.reshape(-1,1)
Sp = dfc.Spain.values.reshape(-1,1)
Ar = dfc.Argentina.values.reshape(-1,1)
Ch = dfc.China.values.reshape(-1,1)
scaled_US = preprocessing.MinMaxScaler().fit(US).transform(US)
scaled_I = preprocessing.MinMaxScaler().fit(I).transform(I)
scaled_Br = preprocessing.MinMaxScaler().fit(Br).transform(Br)
scaled_R = preprocessing.MinMaxScaler().fit(R).transform(R)
scaled_C = preprocessing.MinMaxScaler().fit(B).transform(B)
scaled_Pe = preprocessing.MinMaxScaler().fit(S).transform(S)
scaled_Sp = preprocessing.MinMaxScaler().fit(M).transform(M)
scaled_Ar = preprocessing.MinMaxScaler().fit(Ar).transform(Ar)
scaled_Ch = preprocessing.MinMaxScaler().fit(Ch).transform(Ch)
```

```python
plt.figure(figsize=[10,10])
plt.plot(dfc.index,scaled_US,'-.',color='g',lw=3,label='US')
plt.plot(dfc.index,scaled_I,'-.', color='r',lw=4,label='India')
plt.plot(dfc.index,scaled_Br,'-', color='b',lw=4,label='Brazil')
plt.plot(dfc.index,scaled_R, color='r',lw=4,label='Russia')
plt.plot(dfc.index,scaled_C, '-.',color='k',lw=3,label='Colombia')
plt.plot(dfc.index,scaled_Pe, color='m',lw=3,label='Peru')
plt.plot(dfc.index,scaled_Sp, '-.',color='b',lw=3,label='Spain')
plt.plot(dfc.index,scaled_Ar,'-', color='k',lw=3,label='Argentina')
plt.plot(dfc.index,scaled_Ch,'-', color='g',lw=5,label='China')
plt.legend(fontsize=20)
plt.title('TOP LIST in COVID-19', fontsize=20,weight='bold')
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Date', fontsize=20, weight='bold')
plt.ylabel('Scaled Cases', fontsize=20,weight='bold')
plt.savefig('TOPPER.png')
```
